# Visual EDA: Box Plots

## Box plot

*  A box  plot visualizes how spread out the data points are  away from the center. The box plot has  two parts, the box and the whiskers. 


```python
import seaborn as sns
sns.boxplot(x='day', y='tip', data=tips);
```

#### Commentaries
* *By looking at the graphs, we see that the blue box,  or Sunday, has the widest spread of tip values. On Saturday, we get a lot of outlier  values with a maximum tip value here of 10. Also, just by looking at the  lines bisecting the boxes, we could see that the Sunday and Friday have the  highest median while Thursday has the lowest.*

Let’s validate this by using another  exploratory command: **median**.

```python
tips.groupby(tips.day).median() 
```

We see that  the highest median tip is indeed on Sundays, followed by Friday, and then  Saturday and then Thursday. 
Now let’s review the command and learn  two new commands: groupby and median. Here, groupby is telling  pandas to group the dataframe  by the attribute “day” which we specify by  putting tips.day inside the parenthesis. If you’re familiar with SQL, groupby()  acts like the GROUP BY command in SQL. 

You can also do the group by trick with  other exploratory commands like describe().  

```python
tips.groupby(tips.day).describe() 
```

Now the table  describes the dataset, but organized by day. Okay, now because there are too many  columns, Pandas is truncating the results.  We want to view the tips amounts only, so we  could filter that by editing our code a little. Modify the first part of the code to  this: we change this portion on tips and then we type square brackets and enclose that  with another pair of square brackets. Then inside  the inner most bracket, we type tip in single  quotes. So that’is tip without an s in the end  because we are selecting the column named  “tip”. So this whole thing here tells Pandas,  “in the tip dataset, pull only the tip column.  Group them by the day attribute and then describe  the statistical summaries of the dataset.

```python
tips[['tip']].groupby(tips.day).describe()
```

## Visual EDA: Violin Plots

Now, we are going to level up and use the violin  plot, a richer, more helpful tool for visual EDA. Violin plots are visual aid for visual exploratory data analysis.


```python
# category plot
# sns.set()
sns.catplot(x='day',y='tip',hue='sex',kind='violin', data = tips) # The hue=”sex” parameter tells  seaborn to use different colors  depending on the sex – Male or Female.
```

The legend here says blue is for  Female and this orange color is for Male.  This looks a bit messy -- we have 8 plots  separating the Male and Female data.  
Let’s simplify it by merging the  male and female graphs together.  To do this, we’ll add the  split parameter parenthesis. Let’s insert the split parameter, like this

```python
sns.set_style('whitegrid')
sns.catplot(x="day", y="tip", hue="sex", kind="violin", split=True, height=10, aspect = 11.7/8.27, data=tips);
```

# Exploring Dimensions – Category Plots

In this lesson, we will explore the dataset using various methods available in Seaborn.
In Tableau, we learned that another way to enrich our exploration is by visualizing the effect of other dimensions like **seasonality, profit, and location**.
Today, we will learn about how to visualize dimensions using Jupyter Notebooks, so we could see the data from different perspectives.
To do this, we will use **color**, **size**, **shape**, and **positioning** in the plot.



```python
import pandas as pd, seaborn as sns

tips = pd.read_csv('tips.csv') # Read our CSV file and name it 'tips'
tips # Display the dataframe
```

```python
sns.catplot(data = tips, x='day', y='tip', hue='sex', );
```

Now let’s tweak our visuals  and explore other dimensions. 

```python
sns.catplot(data = tips, x='sex', y='tip', hue='day', );
```

We could see right away there  are more Male customers giving tips and they have the highest ones,  as we saw in the previous visual. It’s a bit hard to distinguish which  day has the most number of tips  but we could actually just scroll up in previous chart to see the plot using day as dimension. Plotting the points using different  dimensions, you could discover more information out of the dataset. Let’s do it again.

```python
sns.catplot(data = tips, x='time', y='tip', hue='sex', );
```

So what we did here is to plot the time of the  meal and that’s just Lunch and Dinner..you could  see that there are more dots at Dinner time. Let’s do one more.

```python
sns.catplot(data = tips, x='time', y='tip', hue='smoker', );
```

Here, at Dinner and Lunch time, it looks  like Non-smokers tend to give higher tips. Here is the maximum tip value of 10 and  it was given by a smoker. If we scroll  
up to the previous visual, we know that this  high tipper-smoker is also Male. So you see  how viewing data from multiple dimensions  creates more insight about the information. 

```python
sns.catplot(data = tips, x='size', y='tip', hue='time', );
```

 What we could see now is that the highest  tip of 10 came from a table of 3. Most of the  other high tip amounts came from a table for 4.  
There seems to be a general trend where the larger  the table size, the larger the tip, of course this  could be because tips are sometimes calculated  as a percent of the total bill, when you have  a larger table size then you will have a larger  total bill. … It seems to be the case from table  size 2, 3, 4, the sizes… they seem to generally  increase in tip size… there’s also this one high  tip from table 6. But at a size of 5, this trend  seems not to apply. This is just from visually  inspecting the plots and you should verify this  using other methods, like regression perhaps. You’ve just learned how to  configure catplots or category plots  to highlight various dimensions in the dataset.
